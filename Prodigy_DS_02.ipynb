Description:- Netflix is a popular streaming service that offers a vast catalog of movies, TV shows, and original contents. This dataset is a cleaned version of the original version which can be found here. The data consist of contents added to Netflix from 2008 to 2021. The oldest content is as old as 1925 and the newest as 2021.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import seaborn as sns
```

```python
data=pd.read_csv('netflix1.csv')
data
```

```python
df = pd.DataFrame(data)
```

```python
# Treat the duplicates
df = df.drop_duplicates()

# Populate missing rows
df.fillna("Unknown", inplace=True)


# Drop unnecessary columns if they exist
if "director" in df.columns:
    df = df.drop(columns=["director"])

print(df)
```

```python
Exploratory Data Analysis (EDA)
```

```python
# Display basic information about the dataset
print("Basic information about the dataset:")
print(df.info())

# Summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
print(df.describe())

```

```python
# Plotting the distribution of 'release_year'
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='release_year', bins=10,color='skyblue')
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()

```

```python
# Plotting the count of each type (Movie vs. TV Show)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='type',color='g')
plt.title('Count of Each Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

```

```python
# Plotting the count of each rating
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order=df['rating'].value_counts().index,color='purple')
plt.title('Count of Each Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

Release Year Distribution:

The distribution of release years shows that the dataset contains movies and TV shows released from as early as 1925 to 2021.
The majority of the content seems to be more recent, with a mean release year of 2014. The distribution appears to be slightly left-skewed, indicating more recent releases.

Type Distribution (Movie vs. TV Show):

The count of each type (Movie vs. TV Show) reveals that there are more movies than TV shows in the dataset.
This suggests that the platform may have a larger collection of movies compared to TV shows.

Rating Distribution:

The count of each rating shows the distribution of content ratings in the dataset.
The most common ratings appear to be TV-MA, TV-PG, and TV-14, indicating that the majority of the content may be suitable for mature audiences.
